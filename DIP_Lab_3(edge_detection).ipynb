```python
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.ndimage.filters import correlate
from scipy.ndimage.filters import convolve
from google.colab.patches import cv2_imshow

from google.colab import drive
```

```python
img = cv.imread('/content/dataset/MyDrive/Lab/powerpuff_girls.jpeg')
plt.imshow(img)
```

**Image Padding.**

```python
img.shape
```

```python
image_padded = np.zeros((img.shape[0] + 2, img.shape[1] + 2))
image_padded[1:-1, 1:-1] = img[:, :, 0]

output = image_padded

plt.imshow(output)
```

**Edge Detection Function.**

```python
def edge(image, kernel):
  for i in range(image.shape[1]):
    for j in range(image.shape[0]):
      output[j, i] = (kernel * image_padded[j: j+3, i: i+3]).sum()
  return output
```

# Prewitt Kernel

```python
p_x = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
```

```python
edge(img, p_x)

# plt.imshow(output)
cv2_imshow(output)
```

```python
p_y = p_x.transpose()

p_y
```

```python
edge(img, p_y)

# plt.imshow(output)
cv2_imshow(output)
```

**Combining vertical and horizontal prewitt kernel :**

```python
p_xy = np.sqrt(p_x*p_x + p_y*p_y)

p_xy
```

```python
edge(img, p_xy)

# plt.imshow(output)
cv2_imshow(output)
```

# Sobel Kernel

# Laplacian Kernel

```python
#when a = 0

a = 0
lap1 = (1/(1+a))*np.array([[a, 1-a, a], [1-a, -4, 1-a], [a, 1-a, a]])
```

```python
edge(img, lap1)

# plt.imshow(output)
cv2_imshow(output)
```

```python
#when a = 0.5

a = 0.5
lap2 = (1/(1+a))*np.array([[a, 1-a, a], [1-a, -4, 1-a], [a, 1-a, a]])
```

```python
edge(img, lap2)

# plt.imshow(output)
cv2_imshow(output)
```
